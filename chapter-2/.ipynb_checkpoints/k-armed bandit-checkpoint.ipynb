```python
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Stationary Average Method

```python
# epsilon-greedy search
def epsilon_greedy(epsilon, steps=1000, epochs=2000, bandits_num=10,
                   initial_q = 0, random_means = np.random.normal(0, 1, 10)):

    std = 1
    optimal_action = np.argmax(random_means)
    avg_eps_reward = np.zeros(steps)
    avg_optimality = np.zeros(steps)

    for e in tqdm(range(epochs)):
        Q = np.zeros(bandits_num)+initial_q
        N = np.zeros(bandits_num)
        rewards_list = np.zeros(steps)
        optimal_action_flg = np.zeros(steps)

        for t in range(steps):
            if np.random.random() >= epsilon:
                max_actions = np.where(Q==np.max(Q))[0]
                action = np.random.choice(max_actions)
            else:
                action = np.random.choice(range(bandits_num))
            reward = np.random.normal(random_means[action], std, 1)
            N[action] += 1
            Q[action] += 1/N[action]*(reward-Q[action]) # update Q-value function
            rewards_list[t] = reward # logging received rewards
            optimal_action_flg[t] = int(action==optimal_action) #logging if a chosen action is optimal
            
        avg_eps_reward += 1/(e+1)*(rewards_list-avg_eps_reward)
        avg_optimality += 1/(e+1)*(optimal_action_flg-avg_optimality)

    return avg_eps_reward, avg_optimality
```

```python
means = np.random.normal(0, 1, 10)
greedy_rewards = epsilon_greedy(epsilon=0, random_means=means)
eps_greedy_rewards = epsilon_greedy(epsilon=0.1, random_means=means)
eps_greedy_rewards2 = epsilon_greedy(epsilon=0.01, random_means=means)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1000), greedy_rewards[0], label='greedy')
plt.plot(range(1000), eps_greedy_rewards[0], label='epsilon=0.1')
plt.plot(range(1000), eps_greedy_rewards2[0], label='epsilon=0.01')
plt.hlines(np.max(means), 0, 1000, linestyles = 'dashed')
plt.xlim(0, 1000)
plt.legend(loc='lower right')
plt.title('Average reward')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1000), greedy_rewards[1], label='greedy')
plt.plot(range(1000), eps_greedy_rewards[1], label='epsilon=0.1')
plt.plot(range(1000), eps_greedy_rewards2[1], label='epsilon=0.01')
plt.hlines(1, 0, 1000, linestyles = 'dashed')
plt.xlim(0, 1000)
plt.legend(loc='lower right')
plt.title('% of optimal action')
plt.show()
```

### 3. Upper-Confidence-Bound Action Selection 

```python
def ucb_action_selection(c, steps=1000, epochs=2000, bandits_num=10,
                          random_means = np.random.normal(0, 1, 10)):

    std = 1
    optimal_action = np.argmax(random_means)
    avg_eps_reward = np.zeros(steps)
    avg_optimality = np.zeros(steps)

    for e in tqdm(range(epochs)):
        Q = np.zeros(bandits_num)
        N = np.zeros(bandits_num)
        rewards_list = np.zeros(steps)
        optimal_action_flg = np.zeros(steps)

        for t in range(steps):
            action = np.argmax(Q+c*np.sqrt(np.log(t+1)/(N+1)))
            reward = np.random.normal(random_means[action], std, 1)
            N[action] += 1
            Q[action] += 1/N[action]*(reward-Q[action]) # update Q-value function
            rewards_list[t] = reward # logging received rewards
            optimal_action_flg[t] = int(action==optimal_action) #logging if a chosen action is optimal
            
        avg_eps_reward += 1/(e+1)*(rewards_list-avg_eps_reward)
        avg_optimality += 1/(e+1)*(optimal_action_flg-avg_optimality)

    return avg_eps_reward, avg_optimality
```

```python
means = np.random.normal(0, 1, 10)
ucb_rewards = ucb_action_selection(c=2, random_means=means)
eps_greedy_rewards = epsilon_greedy(epsilon=0.1, random_means=means)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1000), ucb_rewards[0], label='UCB')
plt.plot(range(1000), eps_greedy_rewards[0], label='eps-greedy')
plt.hlines(np.max(means), 0, 1000, linestyles = 'dashed')
plt.xlim(0, 1000)
plt.legend(loc='lower right')
plt.title('Average reward')
plt.show()
```

### 4. Gradient Bandit Algorithm

```python
def grad_algo(alpha, steps=1000, epochs=2000, bandits_num=10,
                          random_means = np.random.normal(0, 1, 10)):

    std = 1
    optimal_action = np.argmax(random_means)
    avg_eps_reward = np.zeros(steps)
    avg_optimality = np.zeros(steps)
    
    for e in tqdm(range(epochs)):
        R = 0
        n = 0
        H = np.zeros(bandits_num)
        prob = np.exp(H)/np.sum(np.exp(H))
        rewards_list = np.zeros(steps)
        optimal_action_flg = np.zeros(steps)
        
        for t in range(steps):
            action = np.random.choice(range(bandits_num), p=prob)
            reward = np.random.normal(random_means[action], std, 1)
            n += 1
            R += 1/n*(reward-R) # baseline
            H[action] += alpha*(reward-R)*(1-prob[action])
            H[[i for i in range(bandits_num) if i!=action]] -= alpha*(reward-R)*prob[action]
            prob = np.exp(H)/np.sum(np.exp(H))
            rewards_list[t] = reward # logging received rewards
            optimal_action_flg[t] = int(action==optimal_action) #logging if a chosen action is optimal
            
        avg_eps_reward += 1/(e+1)*(rewards_list-avg_eps_reward)
        avg_optimality += 1/(e+1)*(optimal_action_flg-avg_optimality)

    return avg_eps_reward, avg_optimality
```

```python
means = np.random.normal(4, 1, 10)
grad_rewards = grad_algo(alpha=0.1, random_means=means)
grad_rewards2 = grad_algo(alpha=0.4, random_means=means)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1000), grad_rewards[1], label='alpha=0.1')
plt.plot(range(1000), grad_rewards2[1], label='alpha=0.4')
plt.hlines(1, 0, 1000, linestyles = 'dashed')
plt.xlim(0, 1000)
plt.legend(loc='lower right')
plt.title('% of optimal action')
plt.show()
```

### 5. Summary

```python
params = np.logspace(-7, 2, 10, base=2)
```

```python
params
```

```python
means = np.random.normal(0, 1, 10)
learning_curve = np.zeros(len(params[:7]))

for i, x in enumerate(params[:7]):
    rewards = epsilon_greedy(epsilon=x, random_means=means)[0]
    avg_reward = np.mean(rewards)
    learning_curve[i] = avg_reward
```

```python
plt.figure(figsize=(10, 6))
plt.plot(params[:7], learning_curve, label='eps-greedy')
plt.xscale('log', basex=2)
plt.show()
```
